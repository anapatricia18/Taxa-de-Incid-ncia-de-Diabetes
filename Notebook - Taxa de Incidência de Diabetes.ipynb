Primeiramente, vamos determinar qual o maior e menor número de mulheres e homens que sabem ler, para assim conseguirmos trabalhar da melhor forma possível com as dimensões dos gráficos.

```python
max_diab_Homens = df_first["Sexo Masculino"].max() #valor máximo da coluna Sexo_Masculino
min_diab_Homens = df_first["Sexo Masculino"].min() #valor mínimo da coluna Sexo_Masculino

max_diab_Mulheres = df_first["Sexo Feminino"].max() #valor máximo da coluna Sexo_Feminino
min_diab_Mulheres = df_first["Sexo Feminino"].min() #valor mínimo da coluna Sexo_feminino

print("Máximo sexo masculino com diabetes =",max_diab_Homens,";  Mínimo =", min_diab_Homens,";  Máximo sexo feminino com diabetes =", max_diab_Mulheres, ";  Mínimo =",min_diab_Mulheres,";")
```

#### Homens

```python
import numpy as npw
import matplotlib.pyplot as plt #importar da biblioteca dados de gráficos

#dados para o gráfico
Anos = df_first ['Período'].tolist() #transformar os valores da coluna Ano em uma lista numa única variável
homens= df_first['Sexo Masculino'].tolist() #transformar os valores da coluna Sexo_Masculino em uma lista numa única variável

#definições do gráfico
plt.bar(Anos, homens, color="lightblue") #criar um gráfico de barras azul claro com os dados das colunas Anos(x)
plt.gcf().set_size_inches(20, 6) #função para trabalhar as dimensões do gráfico

#estilização do gráfico
plt.xticks(Anos) #título do eixo do x (linha horizontal)
plt.ylabel('Número do sexo masculino') #título do eixo do y (linha vertical)
plt.xlabel('Anos') #valores do eixo do x (todos os valores da coluna Anos)
plt.title('Número de homens por ano com diabetes') #título do gráfico
plt.show() #mostrar o gráfico de barras
```

#### Agora vamos calcular algumas médias importantes para compararmos o conjunto de dados.

```python
media_homens_diab = df_first["Sexo Masculino"].mean() #média de todos os valores da coluna Sexo_Masculino
media_mulheres_diab = df_first["Sexo Feminino"].mean() #média de todos os valores da coluna Sexo_Feminino

media_homens_diab=round(media_homens_diab,2) #arredondar o resultado para 2 casas decimais
media_mulheres_diab=round(media_mulheres_diab,2) #arredondar o resultado para 2 casas decimais

print("Em Portugal, existe em média ",media_homens_diab , "de homens com diabetes e ", media_mulheres_diab, " de mulheres com diabetes entre os anos de 1992 e 2018.")
```

Com isto podemos concluir que entre 1992 e 2018, existem mais homens com diabetes em comparação às mulheres com esta mesma condição.

#### Será que houve alguma alteração entre os anos de 1992 a 2010?

```python
result=df_first.sort_values(by=['Total'])
result_1 = result[(result.Período >= 1992) & (result.Período <= 2010)]
result_1[["Período","Sexo Masculino","Sexo Feminino","Total"]]
```

Com estes dados, foi possível verificar que passados 18 anos houve uma alteração ascendente nos números, ou seja, o algoritmo subiu drasticamente até ao ano de 2010.

#### E será que houve alguma alteração entre os anos de 2010 a 2018?

```python
result=df_first.sort_values(by=['Total'], ascending=False )
result[["Período","Sexo Masculino","Sexo Feminino","Total"]].head(5)
```

Podemos ver que passado 8 anos houve uma alteração significativa e os números foram descendo com o passar dos anos. Apesar de existir progressos desde 2010, houve uma evolução negativa no sexo feminino em 2018.
